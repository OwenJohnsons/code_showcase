#### Part (a) 

Plot a cosine function with frequency 10 Hz, amplitude 1.0 (arbitrary unit) and
phase angle 30◦ in the range t = 0 to 2 seconds (use 500 points). Note: =np.pi

```python
def cosinewave(A, f, t, pa): 
    return A*np.cos(np.pi*t*f + pa)
```

```python
# --- defining variables --- 

t = np.linspace(0, 2, 500)
test = cosinewave(1, 10, t, 30)

# --- Standard Plot Procedure --- 

plt.title("Plot of Cosine Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.plot(test)
plt.grid()
plt.show()
```

#### Part (b)

Hz, amplitude 1.0 and phase angle 30◦ and the other of frequency 8 Hz, amplitude
0.8 and phase angle 60◦.

```python
# --- Plotting two specified waves --- 

plt.plot(cosinewave(1, 10, t, 30), label = "First Wave") 
plt.plot(cosinewave(.8, 8, t, 60), label = "Second Wave")

# --- Standard Plot Procedure --- 

plt.title("Overlay of both waves")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.legend()
plt.grid()
plt.show()
```

Determine and plot the resulting wave that occurs when the two waves from (2b)

```python
# --- Adding two waves to find resultant wave --- 

resultingwave = cosinewave(.8, 8, t, 60) + cosinewave(1, 10, t, 30)

# --- Standard Plot Procedure --- 

plt.plot(resultingwave, label = "Resulting Wave") 

plt.title("Interference of both waves")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.legend()
plt.grid()
plt.show()
```

#### Part (a)

Use the numpy.random.rand() function to generate two arrays containing the
x and y coordinates of 100 virtual raindrops. Note: numpy.random.rand(N)
generates N random numbers from a uniform distribution between 0 and 1, or

```python
# --- Creating two arrays to simulate random raindrops --- 
x = np.random.rand(100)
y = np.random.rand(100)
```

#### Part (b) 

For visualisation plot the raindrops and draw the quarter circle (take care with
the aspect ratio!). Note: keep the plotting separate from the calculation below.

```python
# --- Plotting the quarter circle --- 

theta = np.linspace(0, 1/2*np.pi, 100) # quarter circle so creating 100 points across 90 degree range 

r = np.sqrt(1) # unit circle thus r = 1 

x1 = r*np.cos(theta) # standard trig formuala to create circle 
x2 = r*np.sin(theta) 

fig, ax = plt.subplots(1) #overlaying plots 
ax.plot(x1, x2)
ax.set_aspect(1)

plt.scatter(x, y) #random rain drops 
plt.grid()

plt.show()
```

#### Part (c) 

Determine a logical test to decide whether each raindrop falls within the quarter
circle and use it to count up how many raindrops fall within the quarter circle.

```python
# --- Calculating --- 

def modulus(x, y): 
    return np.sqrt(x**2 + y**2)

raindropdistance = modulus(x, y) # distance of each raindrop from the orgin 

rdwithincircle = raindropdistance[(raindropdistance < 1)] # removes any raindrops outside the boundaries of the unit circle
```

#### Part (d) 

Use the ratio of the number of raindrops in the circle to the total number in the
square to estimate π



```python
piratio = (len(rdwithincircle))/(len(raindropdistance)) # ratio calculation 

print("Esitimated value for pi is", piratio*4) #multiply by four because it's the quarter of a circle 
```

#### Part (e)

Run your program a few times and investigate how the estimate of  changes.
Hint: you could make a function which takes N and returns  to make running
it multiple times and for different values of N easier.

```python
def piratioestimation(N): 
    r = range(N) # number of trials 
    total = 0 # resets total 
    for n in r: 
        x = np.random.rand(100) # generates new raindrops for each trial 
        y = np.random.rand(100)
        raindropdistance = modulus(x, y) # Recalculation of modulus 
        rdwithincircle = raindropdistance[(raindropdistance < 1)] # reduction of raindrops outside cirlce 
        a = (len(rdwithincircle))/(len(raindropdistance))*4
        total += a 
    print('Estimated value of π for', N, 'trials is', np.round(total/N, 5)) # getting average outputted value for pi 
```

```python
piratioestimation(100000) # calling the function 
```

#### Part (f)

Does increasing N help with the accuracy?

###### Answer: 

Yes, increasing the number of trials (N) shows that the estimated value outputted by the ratio tends to the accepted value for pi with each iteration. 

#### Part (g)

Python is slow for loops. Instead of looping over every raindrop the logical test
can be applied to all the raindrops at once using Numpy arrays and the True
values counted with numpy.count_nonzero() function. When you have your
program working using loops try it using Numpy vectorised operations and see if
it is faster (you can time how long it takes to run any cell in a Python Jupyter
notebook by putting %%timeit as the first line in the cell.)

```python
def pi_ratio_numpy(N):
    x = np.random.rand(N)
    y = np.random.rand(N)
    T = np.count_nonzero(modulus(x, y) <= 1) 
    print(T/N*4)
```

```python
%%timeit
pi_ratio_numpy(100000)
```
