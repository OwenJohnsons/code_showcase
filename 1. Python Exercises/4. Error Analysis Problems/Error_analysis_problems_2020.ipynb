```python
d, h = np.loadtxt("volume.dat", unpack = True)

r = 0.5*d 

print("Radius: {:10.3f}".format(np.mean(r)), "Height: {:10.3f}".format(np.mean(h))) #mean values for both radius and diameter
print("Std Radius: {:10.3f}".format(np.std(r)), "Std Height: {:10.3f}".format(np.std(h)))
```

```python
V_1 = np.pi*np.mean(h)*np.mean(r)**2

print("Volume obtained from averaging radius and height,", "{:10.3f}".format(V_1))

r_mean = np.mean(r)
h_mean = np.mean(h)
std_r = np.std(r)/np.sqrt(len(r)) 
std_h = np.std(h)/np.sqrt(len(h))

print("Std Radius: {:10.3f}".format(std_r), "Std Radius: {:10.3f}".format(std_h))

error_1 = V_1*np.sqrt((std_r/r_mean)**2 + (std_h/h_mean)**2)

print("Error: {:10.3f}".format(error_1))
```

```python
V_2 = np.pi*h*r**2

print("Volume obtained from calclating each radius and height value and averaging volumes,", "{:10.3f}".format(np.mean(V_2)))
```

```python
def dh(r): 
    return np.pi*r**2

def dr(r, h): 
    return 2*np.pi*r*h
```

```python
σ_1 = 0.01
σ_2 = 0.01 

σ = np.sqrt(σ_1**2*(dh(r))**2 + σ_2**2*(dr(r, h))**2)

print("{:10.3f}".format(np.mean(σ)))
```
